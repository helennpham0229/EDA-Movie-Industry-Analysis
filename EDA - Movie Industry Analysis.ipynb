## What are the top 5 most profitable genres?

To find out the top 5 most profitable genres, we will need to work with files containing info we may need which are genres, revenues, budgets.

```python
# find out basic info about the movie_budgets file
movie_budgets.info()
```

```python
# production_budget, domestic_gross, and worldwide_gross columns have inappropraite data types
# remove the comma
production_budget = movie_budgets.production_budget.str.replace(",", "")
domestic_gross = movie_budgets.domestic_gross.str.replace(",", "")
worldwide_gross = movie_budgets.worldwide_gross.str.replace(",", "")
production_budget
domestic_gross
worldwide_gross
```

```python
# remove the dollar sign
production_budget = production_budget.str.replace("$", "")
domestic_gross = domestic_gross.str.replace("$", "")
worldwide_gross = worldwide_gross.str.replace("$", "")
production_budget
domestic_gross
worldwide_gross
```

```python
# change data type from object to float
production_budget = production_budget.astype(float)
domestic_gross = domestic_gross.astype(float)
worldwide_gross = worldwide_gross.astype(float)
production_budget
domestic_gross
worldwide_gross
```

```python
# calculate total_gross
total_gross = domestic_gross + worldwide_gross
total_gross
```

```python
# calculate movie_profit
movie_profit = total_gross - production_budget
movie_profit
```

```python
# create movie_budget dataframe 
movie_budget = {"original_title" : movie_budgets["movie"],
                "release_date" : movie_budgets['release_date'],
                "total_gross" : total_gross,
                "movie_profit" : movie_profit}
movie_budget_df= pd.DataFrame(movie_budget)
movie_budget_df
```

```python
# find out basic info about the title_basics file
title_basics.info()
```

```python
# drop null values from "genres" column
title_basics = title_basics.dropna(axis = 0, how = 'any', thresh = None, subset = ["genres"], inplace = False)
title_basics
```

```python
# merge 2 dataframes "movie_budget_df" and "title_basics"
movie_budget_title = pd.merge(title_basics, movie_budget_df, how = 'inner', on = 'original_title')
movie_budget_title
```

```python
# create dataframe to sort values of the top 200 profitable movies in descending order
cleaned_movie_budget_title = movie_budget_title.sort_values("movie_profit", ascending=False).head(200)
cleaned_movie_budget_title
```

```python
# create function to count number of movies in each genre
def count_genre_number (data,column):
    genre_number = {}
    for genres in data[column]:
        for genre in genres.split(","):
            if genre in genre_number:
                genre_number[genre] += 1
            else:
                genre_number[genre] = 1
    return genre_number   
```

```python
# use the function created to count number of movies in each genre in the combined dataframe
count_of_genre = count_genre_number (cleaned_movie_budget_title,"genres")   
print(count_of_genre)
```

```python
# create list to contain elements for x and y values of the bar chart
x_genres = []
y_num_of_movies = []

for genre, count in count_of_genre.items():
    x_genres.append(genre)
    y_num_of_movies.append(count)
```

```python
# plot bar chart to show each genre vs. number of movies in each genre
plt.figure(figsize = (15, 8))
sns.barplot(x = x_genres, y = y_num_of_movies, color = "blue")
plt.title("Genres from the top 200 most profitable movies", fontsize = 18)
plt.ylabel("Number of movies in each genre", fontsize = 13)
plt.xlabel("Genres", fontsize = 13)
plt.xticks(rotation = 90)
plt.show()
```

**Observing the bar chart above, from the top 200 most profitable movies, our company should invest in Adventure, Action, Comedy, Animation, and Sci-Fi movies**

## When is the best time to release movies?

To find this out, we will need to work with files containing info we may need which are release dates, revenues, budgets.

```python
# revisit movie_budgets
movie_budgets.info()
```

```python
# Convert the release_date field to type datetime object
movie_budget_df['release_date'] = pd.to_datetime(movie_budget_df['release_date'])
```

```python
# Add a new column called release_month with only the month from the release_date
month =  [x.strftime('%B') for x in movie_budget_df['release_date']]
movie_budget_df['release_month'] = month
```

```python
# Caclulate profit by month
profit_vs_month = movie_budget_df.groupby('release_month', as_index=False)['movie_profit'].median().sort_values(by='movie_profit', 
                                                                                                                ascending=False)
profit_vs_month
```

```python
profit_vs_month['release_month'] = pd.Categorical(profit_vs_month['release_month'], ['January', 'February', 'March', 
                                                                                     'April', 'May', 'June', 'July', 
                                                                                     'August', 'September', 'October', 
                                                                                     'November', 'December'])
profit_by_month = profit_vs_month.sort_values("release_month")
profit_by_month
```

```python
# Plot profit_vs_month
plt.figure(figsize = (13, 6))
sns.lineplot(x = profit_by_month['release_month'],
            y = profit_by_month['movie_profit'],
            color = 'red')
plt.title("Movie Profit vs. Release Month", fontsize = 18)
plt.ylabel("Movie Profit (Tens of Millions)", fontsize = 13)
plt.xlabel("Months", fontsize = 13)
plt.show()
```

**Based on the line plot above, June, July, and November are the most profitable months to release movies.**

```python
director_profit = pd.DataFrame(cleaned_profit_director_df.groupby('director_name')['movie_profit'].sum())
director_profit
```

```python
best_directors_df = count_directors.merge(director_profit, 
                                         left_on = count_directors.index, 
                                         right_on = director_profit.index, 
                                         how = 'outer')
best_directors_df = best_directors_df.sort_values(by = 'movie_profit', ascending = False)
```

```python
# Plot net profit vs. count of movies directed 
plt.figure(figsize = (15, 8))
sns.boxplot(best_directors_df.original_title, best_directors_df.movie_profit, color = 'orange')
plt.title('Movie Profit vs. Movies Directed', fontsize = 18)
plt.xlabel('Number of Movies Directed', fontsize = 13)
plt.ylabel('Movie Profit (in Billions)', fontsize = 13)
plt.show()
```

**Looks like directors who directed 4 or more films would make positive net profit.**

```python
# Filter dataframe with 4 or more movies, top 20 by net profit
directed_at_least_4 = best_directors_df[(best_directors_df['original_title'] >= 4) & 
                                        (best_directors_df['movie_profit'] > 0)][:20]
# Find the mean_profit for each of directors who directed 4 or more movies
directed_at_least_4['mean_profit'] = directed_at_least_4.movie_profit / directed_at_least_4.original_title
directed_at_least_4 = directed_at_least_4.sort_values(by = 'mean_profit', ascending = False)
directed_at_least_4
```

```python
cleaned_directed_at_least_4 = {'director_name' : directed_at_least_4['key_0'],
                              'mean_profit' : directed_at_least_4['mean_profit']}
director_mean_profit_df = pd.DataFrame(cleaned_directed_at_least_4)
director_mean_profit_df
```

```python
plt.figure(figsize = (15, 9))
sns.barplot(director_mean_profit_df.director_name, director_mean_profit_df.mean_profit, palette='mako')
plt.xlabel('Top 20 Directors', fontsize = 13)
plt.xticks(rotation = 90)
plt.ylabel('Average Profit (hundreds of millions)', fontsize = 13)
plt.title('Top 20 Directors', fontsize = 18)
plt.show()
```

**We shoud hire Joss Whedon, Christopher Nolan, and/or Jon Favreau.**

## Who are our biggest competitors?

To find this out, we would need to know studio, budget, revenue

```python
movie_gross.info()
```

```python
# remove comma
movie_gross.foreign_gross = movie_gross.foreign_gross.str.replace(",","")
movie_gross.foreign_gross
```

```python
# fill NaN with 0
movie_gross.foreign_gross = movie_gross.foreign_gross.fillna(0)
movie_gross.foreign_gross
```

```python
# change data type
movie_gross.foreign_gross = movie_gross.foreign_gross.astype(float)
movie_gross.foreign_gross
```

```python
# find total_gross
total_gross = movie_gross.domestic_gross + movie_gross.foreign_gross
total_gross
```

```python
# create dataframe 
movie_gross_clean ={"title" : movie_gross["title"],
                    "studio" : movie_gross["studio"],
                    "domestic_gross" : movie_gross["domestic_gross"],
                    "foreign_gross" : movie_gross["foreign_gross"],
                    "total_gross" : total_gross}
movie_gross_clean_df = pd.DataFrame(movie_gross_clean)
movie_gross_clean_df
```

```python
movie_gross_clean_df = movie_gross_clean_df.dropna(axis=0, how='any', thresh=None, subset=["studio"], inplace=False)
movie_gross_clean_df
```

```python
final_movie_gross_clean = movie_gross_clean_df.sort_values("total_gross", ascending=False).head(50)
final_movie_gross_clean
```

```python
def count_studio(dataset,column):
    studio_number={}
    for studios in dataset[column]:
        for studio in studios.split(","):
            if studio in studio_number:
                studio_number[studio]+=1
            else:
                studio_number[studio]=1
    return studio_number
```

```python
final_studio_count=count_studio(final_movie_gross_clean,"studio")   
print(final_studio_count)
```

```python
x_studios = []
y_count= []

for studio, count in final_studio_count.items():
    x_studios.append(studio)
    y_count.append(count)
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x_studios,y_count, palette = "rocket")
plt.title("Studios from the top 50 profitable movies", fontsize = 18)
plt.ylabel("Number of movies released", fontsize = 13)
plt.xlabel("Studios", fontsize = 13)
plt.show()
```

**From the top 50 most profitable films, BV and WB are our biggest competitors.**
